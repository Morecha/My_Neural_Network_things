```python
def bin_enc(lbl):
    mi = min(lbl)
    length = len(bin(max(lbl) - mi + 1)[2:])
    enc = []

    for i in lbl:
        b = bin(i - mi)[2:].zfill(length)
        enc.append([int(n) for n in b])
    return enc

def bin_dec(enc, mi=0):
    lbl = []

    for e in enc:
        rounded = [int(round(x)) for x in e]
        string = ''.join(str(x) for x in rounded)
        num = int(string, 2) + mi
        lbl.append(num)
    return lbl
```

```python
labels = 1, 2, 3, 4
enc = bin_enc(labels)
dec = bin_dec(enc, min(labels))
print(enc)
print(dec)
```

```python
import numpy as np
def onehot_enc(lbl, min_val=0):
    mi = min(lbl)
    enc = np.full((len(lbl), max(lbl) - mi + 1), min_val, np.int8)
    for i, x in enumerate(lbl):
        enc[i, x - mi] = 1
    return enc
    
def onehot_dec(enc, mi=0):
    return [np.argmax(e) + mi for e in enc]
```

```python
labels = 1, 2, 3, 4
enc = onehot_enc(labels)
dec = onehot_dec(enc, min(labels))
print(enc)
print(dec)
```

```python
def sig(X):
  return [1 / (1 + np.exp(-x)) for x in X]
def sigd(X):
  output = []
  for i, x in enumerate(X):
    s = sig([x])[0]
    output.append(s * (1 - s))
  return output
```

```python
import pandas as pd

data = pd.read_csv('iris_csv.csv')
data.head()
```

```python
#read class column from variable data
kelas = list(data.iloc[:, -1])
kelas
```

```python
def bin_enc_str(kelas):
    classes, uniques = pd.factorize(kelas)  #memisah fungsi kelas dan jenis
    mi = min(classes)       #mencari nilai minimum
    length = len(bin(max(classes) - mi + 1)[2:])    #mencari panjang biner
    enc = []    
    for i in classes: 
        b = bin(i - mi)[2:].zfill(length)       #mengubah ke bentuk biner dan menambahkan angka 0 didepannya
        enc.append([int(n) for n in b])         #mengubah ke bentuk list memasukkanya ke variable nec
    return enc, uniques

def bin_dec_str(enc, uniques, mi=0):
    classes = []
    class_label = []
    for e in enc:
        rounded = [int(round(x)) for x in e]
        string = ''.join(str(x) for x in rounded)
        num = int(string, 2) + mi
        classes.append(num)
    for i in range(len(classes)):
        class_label.append(pd.factorize(uniques)[1][classes[i]])
    return class_label
```

```python
enc, uniques = bin_enc_str(kelas)
dec = bin_dec_str(enc, uniques)
print(enc)
print(dec)
```

```python
def oneshot_enc_str(kelas, min_val=0):
    classes, uniques = pd.factorize(kelas)
    mi = min(classes)
    enc = np.full((len(classes), max(classes) - mi + 1), min_val, np.int8)
    for i, x in enumerate(classes):
        enc[i, x - mi] = 1
    return enc, uniques


def oneshot_dec_str(enc, uniques, mi=0):
    classes = []
    classes_label = []
    for e in enc:
        index_of = np.where(e==1)[0]
        classes_label.append(uniques[index_of])
    return classes_label
```

```python
enc, uniques = oneshot_enc_str(kelas)
dec = oneshot_dec_str(enc, uniques)
print(enc)
print(dec)
```
